```python
import cv2
import matplotlib.pyplot as plt
import matplotlib
```

```python
plt.figure(figsize=(20,10))
image = cv2.imread('batman_aspal.png')
plt.imshow(image[...,::-1])
```

```python
plt.figure(figsize=(20,10))
image = cv2.imread('batman_aspal.png')
# Batman Asli
cv2.rectangle(image, (140,80),(252,200),(0,0,255),3)
cv2.rectangle(image, (138,60),(254,80),(0,0,255),-1)
cv2.putText(image,'Batman Asli', (140,75), cv2.FONT_HERSHEY_COMPLEX, 0.5, (255,255,255), 1, cv2.LINE_AA)
# Batman KW
cv2.rectangle(image, (495,116), (608,235),(0,0,255),3)
cv2.rectangle(image, (493,94), (610,116),(0,0,255),-1)
cv2.putText(image,'Batman KW', (495,110), cv2.FONT_HERSHEY_COMPLEX, 0.5, (255,255,255), 1, cv2.LINE_AA)
plt.imshow(image[...,::-1])
```

```python
cv2.imwrite('batman_annot.png',image)
```
